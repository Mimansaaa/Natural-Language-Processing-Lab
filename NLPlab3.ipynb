```python
from collections import Counter
with open('textdata_NLP.txt', 'r') as file:
  content=file.read()

print(content)
words=content.split()

frequency=Counter(words)

for word, frequency in frequency.items():
    print(f'{word}: {frequency}')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
cloud = WordCloud(width=800, height=400, background_color='white', colormap="brg").generate(content)

plt.figure(figsize=(10, 5))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import re
words = re.findall(r'\b\w+\b', content.lower())

frequency = Counter(words)
top_10_words = frequency.most_common(10)

for word, frequency in top_10_words:
    print(f'{word}: {frequency}')
```

```python
import matplotlib.pyplot as plt
top_words, frequencies = zip(*top_10_words)

plt.figure(figsize=(10, 5))
plt.bar(top_words, frequencies, color='lightblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 occuring words')
plt.show()
```
